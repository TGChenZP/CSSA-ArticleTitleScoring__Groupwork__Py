```python
# open the file Title Scoring Results.xlsx and read the data
import pandas as pd
df = pd.read_excel('Title Scoring Results.xlsx')
```

```python
# select the columns to be displayed
selected_columns = ["Model", "Embedding", "Features", "Val R2", "Test R2"]
selected_data = df[selected_columns]
```

```python
#
markdown_table = selected_data.to_markdown(index=False)
```

```python
# Save the DataFrame as a markdown file
with open('Title Scoring Results.md', 'w') as f:
    f.write(markdown_table)

print(markdown_table)
```
